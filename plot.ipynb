```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Data for the first plot (Breast Cancer)
data_cancer = {
    'Qubit': [4, 5, 6, 7],
    'GA-QSVM': [91.5, 93.2, 94.2, 92.6],
    'Z-Feature': [75.5, 77.0, 76.7, 77.5],
    'RBF': [94.5, 95.2, 94.7, 95.0]
}
df_cancer = pd.DataFrame(data_cancer).set_index('Qubit')

# Data for the second plot (Digits)
data_digits = {
    'Qubit': [4, 5, 6, 7],
    'GA-QSVM': [76.5, 79.1, 83.1, 84.5],
    'Z-Feature': [54.7, 58.3, 63.5, 67.0],
    'RBF': [78.3, 82.6, 84.7, 87.3]
}
df_digits = pd.DataFrame(data_digits).set_index('Qubit')

# Data for the third plot (Wine)
data_wine = {
    'Qubit': [4, 5, 6, 7],
    'GA-QSVM': [96.41, 97.05, 97.44, 97.43],
    'Z-Feature': [90.51, 90.51, 90.9, 91.15],
    'RBF': [96.54, 96.41, 97.18, 97.69]
}
df_wine = pd.DataFrame(data_wine).set_index('Qubit')

# Set seaborn style
sns.set_theme(style="whitegrid")

# Create a figure with 3 subplots, sharing the x-axis but NOT the y-axis
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharex=True)

# --- Plot 1: Breast Cancer ---
sns.lineplot(data=df_cancer, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[0], legend=False)
axes[0].set_title('Breast Cancer Dataset', fontsize=16)
axes[0].set_ylabel('Accuracy (%)', fontsize=12)

# --- Plot 2: Digits ---
sns.lineplot(data=df_digits, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[1], legend=False)
axes[1].set_title('Digits Dataset', fontsize=16)
axes[1].set_xlabel('Number of Qubits', fontsize=14)
axes[1].set_ylabel('') # Remove redundant y-axis label

# --- Plot 3: Wine ---
# We plot with a legend here so we can grab the handles and labels from it
sns.lineplot(data=df_wine, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[2])
axes[2].set_title('Wine Dataset', fontsize=16)
axes[2].set_ylabel('') # Remove redundant y-axis label

# Set integer ticks for the shared x-axis
plt.xticks(ticks=df_cancer.index)

# Create a single shared legend from the last plot (which has the legend data)
handles, labels = axes[2].get_legend_handles_labels()
axes[2].get_legend().remove() # Remove the individual legend from the third plot
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.05), ncol=3, fontsize=12, title="Algorithm")

# Adjust layout to prevent titles and legend from overlapping
plt.tight_layout()
plt.subplots_adjust(bottom=0.2) # Make space for the bottom legend

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Data Extraction ---

# Data for the Wine Dataset
data_wine = {
    'Qubit': list(range(3, 8)),
    'RBF': [91, 91, 94, 94, 92],
    'ZZ-Feature': [91, 94, 94, 92, 91],
    'GA-QSVM': [95, 95, 97, 99, 97]
}
df_wine = pd.DataFrame(data_wine).set_index('Qubit')

# Data for the Breast Cancer Dataset
data_cancer = {
    'Qubit': list(range(3, 8)),
    'RBF': [87, 91, 90, 90, 90],
    'ZZ-Feature': [83, 85, 81, 83, 71],
    'GA-QSVM': [91, 94, 95, 96, 97]
}
df_cancer = pd.DataFrame(data_cancer).set_index('Qubit')

# Data for the Digits Dataset
data_digits = {
    'Qubit': list(range(3, 8)),
    'RBF': [64, 72, 74, 80, 77],
    'ZZ-Feature': [24, 28, 24, 37, 42],
    'GA-QSVM': [64, 75, 77, 81, 82]
}
df_digits = pd.DataFrame(data_digits).set_index('Qubit')


# --- Plotting ---

# Set seaborn style
sns.set_theme(style="whitegrid")

# Create a figure with 3 subplots, sharing the x-axis but NOT the y-axis
fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharex=True)

# --- Plot 1: Wine ---
sns.lineplot(data=df_wine, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[0], legend=False)
axes[0].set_title('Wine Dataset', fontsize=16)
axes[0].set_ylabel('Accuracy (%)', fontsize=12)

# --- Plot 2: Breast Cancer ---
sns.lineplot(data=df_cancer, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[1], legend=False)
axes[1].set_title('Breast Cancer Dataset', fontsize=16)
axes[1].set_xlabel('Number of Qubits (PCA)', fontsize=14)
axes[1].set_ylabel('') # Remove redundant y-axis label

# --- Plot 3: Digits ---
# Plot with a legend to grab handles and labels
sns.lineplot(data=df_digits, marker='o', linewidth=2.5, markersize=8, dashes=False, ax=axes[2])
axes[2].set_title('Digits Dataset', fontsize=16)
axes[2].set_ylabel('') # Remove redundant y-axis label

# --- Final Touches ---

# Set integer ticks for the shared x-axis to cover the full range
plt.xticks(ticks=df_wine.index)

# Create a single shared legend
handles, labels_orig = axes[2].get_legend_handles_labels()
# Rename labels for consistency
labels = [l.replace('ZZ-Feature', 'ZZ') for l in labels_orig]
labels = [l.replace('GA-QSVM', 'Genetic Algorithm') for l in labels]

axes[2].get_legend().remove() # Remove the individual legend
fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5, -0.08), ncol=3, fontsize=12, title="Feature Map / Algorithm")

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(bottom=0.2) 

plt.show()
```
